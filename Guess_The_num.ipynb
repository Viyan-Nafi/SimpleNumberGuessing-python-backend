```python
import random

def play_again():
    value = input("Do you want to play again?? Enter yes or no ").strip().lower()
                
    while value not in ['yes', 'no']:
        value = input("Please enter a valid value! Enter yes or no ").strip().lower()
                    
    return value
        
def guess_num():
    
    while True:

        start = input("Welcome to 'Guess-the-number' game! Do you want to play?? Enter yes or no ").strip().lower()
        while start not in ['yes', 'no']:
            start = input("Please enter a valid value! Enter yes or no only! ").strip().lower()
            
        if start == 'no':
            print("Thank you!")
            break
        
        
        r_num = random.randint(1,10) # stores a random number
        num = ''
        while num.isdigit() == False or int(num) not in range(1,11):
            num = input("Please enter any number between 1 and 10! ").strip() # text
        
        while int(num) != r_num:
            num = input("Your guess is wrong:( Please try again! ").strip() # text
            
        print("Your guess is right:)")
        value = play_again()
                
        if value == 'yes':  
            continue
                    
        elif value == 'no':
            print("Thank you!")
            break
                
                
            
                
guess_num()  #Run the game :)          
        
```
